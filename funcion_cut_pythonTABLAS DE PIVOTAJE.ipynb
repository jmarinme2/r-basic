# Tablas de pivotaje

```python
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
titanic = sb.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby('sex').mean()
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
titanic.groupby(['sex', 'pclass'])[['survived']].aggregate('mean').unstack()
```

El equivalente a todo ese código, que es: agrupame por sexo y clase (groupby([sex, pclass]) y calculame la media (mean()) de los supervivientes ([survived]) de una manera no apilada (unstack), es el emplear una tabla de pivotaje, que es un método de todos los dataframes:
En el invocamos fd.pivot_table(columna de interés para el cálculo, index=agrupado por lo que quiero que aparezca en las filas, columns=agrupado por lo que quiero que aparezca en las columnas)
El argumento aggfunct, es donde podemos colocar la función que nos interesa, por defecto está np.mean. **SI QUIERO QUE HAGA VARIAS FUNCIONES LO TENGO QUE PONER EN UN DICCIONARIO.**

```python
#el equivalente a todo el código de la celda anterior es la tabla de pivotaje
titanic.pivot_table('survived', index='sex',columns='pclass', aggfunc= np.mean, margins=True) 
```

Podemos transformar una variable cuantitativa en categórica empleando la fucnión cut() 

```python
age = pd.cut(titanic['age'], bins=[0,18,25,35,50,65,100], right=False, include_lowest=True, precision=1)
age
```

```python
titanic.pivot_table('survived', index=['sex', age], columns='class', aggfunc=sum)
```

```python
titanic.pivot_table('survived', index=['sex', age], columns='class')
```

```python
fare = pd.qcut(titanic['fare'], q = 4, labels=['Q1','Q2','Q3','Q4'])
fare
```

```python
 titanic.pivot_table('survived', index=['sex', age], columns=[fare, 'pclass'])
```

```python
titanic.pivot_table(index='sex', columns='class', aggfunc={'survived': sum, 'fare':'mean' })
```

```python
titanic.pivot_table('survived', index='sex', columns='class', margins=True, margins_name='Total')
```
